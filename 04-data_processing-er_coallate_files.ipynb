```python
import os
import glob
import shutil

import pandas as pd
```

```python
wrk_dir=os.getcwd()
os.chdir(f"{wrk_dir}/data/old/raw")
file_lst=[f"{os.getcwd()}/{file}" for file in glob.glob("*.csv")]
os.chdir(wrk_dir)
```

```python
df = pd.read_csv(file_lst[0], index_col=0)
```

```python
df
```

```python
df = df.drop(["source","currYr"], axis=1).melt(id_vars=["year", "label"])
```

```python
df_lst = []
for file in file_lst:
    df_lst.append(pd.read_csv(file, index_col=0))

    df_lst[-1] = df_lst[-1].drop(["source","currYr"], axis=1)
    df_lst[-1]["file_name"] = file.rsplit("/")[-1].split(".")[0]    
```

```python
all_df = pd.concat(df_lst)
```

```python
all_df["label"] = all_df["label"].str.lower()
```
