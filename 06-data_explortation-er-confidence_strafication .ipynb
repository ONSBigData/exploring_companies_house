in previous checks, documents with a low confidence variable, often have errors, though the variable with the error is not low

```python
to_sample = pd.qcut(df.groupby("file_name").conf.min(),
                    q=[0 , .25 , .5 , .75 ,1],
                    labels=[1,2,3,4],
                    duplicates='drop',
                    retbins=False
                   ).reset_index()
```

```python
sample_lst = pd.concat([to_sample[to_sample.conf==n].sample(n=10, random_state=42) 
                        for n in range(1,5)
                       ])
```

```python
source = f"{os.getcwd()}/data/old/raw/"
dest = f"{os.getcwd()}/data/old/conf_check/"
for file in sample_lst.file_name.to_list():
    try:
        shutil.copy(f"{source}{file}.pdf", dest)
    except:
        #some are .html
        shutil.copy(f"{source}{file}.html", dest)
```

```python
pd.merge(df, sample_lst.rename(columns={"conf":"conf_quart"}), on="file_name").to_csv(f"{dest}to_check.csv")
```
