```python
def find_balance_sheet_pages_v2(df):
    
    agg_lines = xip.aggregate_sentences_over_lines(df)
    balance_sheet_mask = ["balancesheet" in text for text in agg_lines["text"]]
    statement_mask =  ["statementoffin" in text for text in agg_lines["text"]]
    notes_mask = ["notestothefinancialstatements"  in text for text in agg_lines["text"]]
    changes_mask = ["Statement of changes in equity" in text for text in agg_lines["text"]]
    
    # we inclulude pages that have balance sheet or 
    asset_sheet_mask = [(balance | statement) and not(notes | changes) 
                        for balance, statement, notes, changes in zip(balance_sheet_mask, 
                                                                      statement_mask,
                                                                      notes_mask,
                                                                      changes_mask
                                                                     )]
    return agg_lines[asset_sheet_mask].csv_num.unique().tolist()
```

```python
new_balance_pages = [find_balance_sheets_v2(data) 
                     for data in tqdm_notebook(no_balance_data_lst)]
```
